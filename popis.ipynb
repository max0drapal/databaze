1. Analýza atributu
Úkol: Jaká data se v něm nacházejí a jakou mají strukturu?

Odpověď:
Ve sloupci description jsou uložena data ve formátu JSON (JavaScript Object Notation).

Struktura: Jedná se o nestrukturovaný/polo-strukturovaný objekt (key-value páry).

Obsah: Data obsahují technické specifikace produktů, které se liší podle typu zboží (např. u elektroniky brand, cpu, ram_gb; u nábytku ergonomic, color).

Klíčový prvek: Společným prvkem pro všechny záznamy je klíč "category", který definuje typ produktu.

2. Analýza SELECT dotazu
Úkol: Co provede dotaz:

```python
SELECT
    name,
    JSON_VALUE(description, '$.category') AS category
FROM product;
```

Odpověď:
Tento dotaz vybere název produktu (name) a k němu přidá virtuální sloupec pojmenovaný category. Hodnota v tomto sloupci je extrahována přímo z JSON dokumentu ve sloupci description pomocí cesty $.category (znak $ reprezentuje kořen dokumentu).

3. SQL skripty pro konkrétní zadání
3.1 Zobrazení všech produktů z vybraného skladu
(Příklad pro sklad s ID 1 – Central Warehouse)

```python
-- Výběr produktů podle cizího klíče warehouse_id
SELECT * FROM product 
WHERE warehouse_id = 1;
```

3.2 Produkty, které mají v kategorii uvedeno "cable"

```python
-- Filtrování pomocí extrakce hodnoty z JSONu
SELECT * FROM product 
WHERE JSON_EXTRACT(description, '$.category') = 'cable';

-- V moderním MySQL lze použít i zkrácený operátor:
-- SELECT * FROM product WHERE description->>'$.category' = 'cable';
```

3.3 Produkty s počtem > 20 seřazené vzestupně

```python
-- Filtrování numerické hodnoty a řazení
SELECT name, quantity 
FROM product 
WHERE quantity > 20 
ORDER BY quantity ASC;
```

3.4 Produkty uskladněné v březnu 2024

```python
-- Filtrování podle rozsahu datumu v březnu
SELECT name, stored_from 
FROM product 
WHERE stored_from >= '2024-03-01' AND stored_from <= '2024-03-31';

-- Alternativní zápis pomocí funkce MONTH a YEAR:
-- SELECT * FROM product WHERE MONTH(stored_from) = 3 AND YEAR(stored_from) = 2024;
```
